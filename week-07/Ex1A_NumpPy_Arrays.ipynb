# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

```python
## Looping Through Arrays
```

```python
arrA = np.array([
    [11, 12, 13,],
    [14, 15, 16,]
    ])

print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

The key difference between these two results are the different dimensions.  
In the first result, we are only looping through the first dimension, aka the rows kind of like:  
2*(11, 12, 13)
2*(14, 15, 16)
 The second result is going through the second dimension, the individual values within the rows, which is like:  
2*(11),  
2*(12),  
2*(13)...

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)

# This result is different because loop_result is representing the entire row, so when you *2, the row is printed repeatedly.
```

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0) #if left empty,axis=0 is the default value

print(joined_array_1)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1)) #axis=0

print(stacked_array_1)
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)

print(splitsky)

# There are now 2 arrays, and each array has two dimensions
```

```python
splitterino = np.array_split(joined_array_1, 3)

print(splitterino)

#There are now 3 arrays each with still two dimensions
```
