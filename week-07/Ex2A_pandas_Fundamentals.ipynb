## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = {9, 10, 11, 12}
ad = {
    'first'   : 13, 
    'second'  : 14, 
    'third'   : 15, 
    'fourth'  : 16
}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(list(ac)) #Casts ac as a list because creating a series with a set doesn't work, as a series uses the index to order and identify each value, and sets aren't ordered. 
ser4 = pd.Series(ad)

print(ser1)
print(ser1)
print(ser3)
print(ser4)
```
