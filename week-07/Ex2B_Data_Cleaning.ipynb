```python
import numpy as np
import pandas as pd
```

```python
df.describe()
```

## Filtering a DataFrame

```python
df_18up = df[(df['Age'] > 18)]
df_18up
```

```python
df_51to99 = df[(df['Score'] > 50) & (df['Score'] < 100)] #Remember to use '&' for AND in conditional logic

df_51to99
```

## Modifying a DataFrame

```python
df_cleaned = df

for x in df.index:
    if df.loc[x, 'Score'] < 0:
        df.drop(x, inplace=True)

df_cleaned
```

```python
for x in df.index:
    if df.loc[x, 'City'] == 'City':
        df.drop(x, inplace=True)

df_cleaned
```

```python
df_cleaned.drop(columns=['City'], inplace=True)
df_cleaned
```

```python
#Can use fillna.df_cleaned['Age].mean() instead of fillna.(25) from the describe info above.
df_cleaned['Age'] = df_cleaned['Age'].fillna(25)

df_cleaned
```

```python
df_cleaned['Score'] = df_cleaned['Score'].replace(0, 66.3)

df_cleaned
```

```python
df_cleaned = df_cleaned.dropna()

df_cleaned
```
