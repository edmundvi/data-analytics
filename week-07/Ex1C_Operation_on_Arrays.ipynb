## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=(2))
print(f'r_1: \n {r_1}')

r_2 = np.random.randint(1, 20, size=(1, 2))
print(f'r_2: \n {r_2}')

r_3 = np.random.rand(2, 1)
print(f'r_3: \n {r_3}')

r_4 = np.random.randint(20, 100, size=(4, 5))
print(f'r_4: \n {r_4}')
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)

print(r_4)
```

The elements in the array were shuffled as whole rows. The entire rows shuffled but not the values within them.

```python
new_shuffle = np.random.permutation(r_4)

print(new_shuffle)
print('\n')
print(r_4)
```

Shuffle randomizes the currenty array, while permutation creates a new one in a random order.  
    Shuffle is useful if you will be working with the same array  
    while permutation is useful to create multiple new randomized arrays.

## Rounding

```python
print(r_3)
```

```python
print(np.trunc(r_3)) #trunc chops off decimals, bascially rounds down
```

```python
print(np.fix(r_3)) # fix rounds to nearest integer towards zero
```

```python
print(np.floor(r_3)) # floor rounds down
```

```python
print(np.ceil(r_3)) # ceil rounds up
```

```python
print(np.round(r_3, 2))
```

## Grocery Problem

```python
#arrays are ordered as vegetables, fruits, boxes of snacks, beverages
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
print(f'The first bag contains {np.sum(bag1)} items, and the second bag contains {np.sum(bag2)} items.')
```

```python
bagsum = bag1 + bag2

print(f'There are {bagsum[0]} vegetables, {bagsum[1]} fruits, {bagsum[2]} boxes of snacks, and {bagsum[3]} beverages in the two grocery bags')
```

## Teamwork

```python
#values are ordered as: mon, tue, wed, thur, fri
Ali = np.array([5, 4, 6, 4, 2])
```

```python
Ben = np.array([0, 5, 5, 3, 4])
```

```python
Carla = np.array([3, 6, 0, 3, 4])
```

```python
all_knocks = np.array([
    Ali,
    Ben,
    Carla
])

print(all_knocks)
```

```python
#Ali 21, Ben 17, Carla 16
print(np.sum(all_knocks, axis=1))

#sum of all doors knocked on = 54
print(np.sum(all_knocks))
```

```python
#sum of all knocks for each day, turns the 3 rows into 1
daily_knocks = np.sum(all_knocks, axis=0)

#find cumulative sum of each day, and multiply by 0.5 to convert the knocks into votes
cum_votes= np.cumsum((daily_knocks * 0.5))
print(f'Rolling total of votes for each day:\n{cum_votes}')

```
