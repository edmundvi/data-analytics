## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=(2))
print(f'r_1: \n {r_1}')

r_2 = np.random.randint(1, 20, size=(1, 2))
print(f'r_2: \n {r_2}')

r_3 = np.random.rand(2, 1)
print(f'r_3: \n {r_3}')

r_4 = np.random.randint(20, 100, size=(4, 5))
print(f'r_4: \n {r_4}')
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)

print(r_4)
```

The elements in the array were shuffled as whole rows. The entire rows shuffled but not the values within them.

```python
new_shuffle = np.random.permutation(r_4)

print(new_shuffle)
print('\n')
print(r_4)
```

Shuffle randomizes the currenty array, while permutation creates a new one in a random order.  
    Shuffle is useful if you will be working with the same array  
    while permutation is useful to create multiple new randomized arrays.

## Mathematical Operations

```python
print('r_1 and r_2:')
print(r_1)
print(r_2)

print('\n shapes:')
print(r_1.shape)
print(r_2.shape)

print('\n r_1 and r_2 added together plus shape:')
print(np.add(r_1, r_2))
print(np.add(r_1, r_2).shape)
```

```python
print('r_1 and r_3:')
print(f'{r_1}\n{r_3}')

print('\n shapes:')
print(f'{r_1.shape}\n{r_3.shape}')

print('\n r_1 and r_3 added together plus shape:')
print(np.add(r_1, r_3))
print(np.add(r_1, r_3).shape)
```

```python
print('r_2 and r_4:')
print(f'{r_2}\n{r_4}')

print('\n shapes:')
print(f'{r_2.shape}\n{r_4.shape}')

print('\n r_2 and r_4 added together plus shape:') # See code below for explanation of error
print(np.add(r_2, r_4))
print(np.add(r_2, r_4).shape)
```
