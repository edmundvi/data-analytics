# Exercise 3.A Getting Familiar with Matplotlib

```python

import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Top of graph!')
plt.xlabel('Bottom of graph')
plt.ylabel('Left of graph')
plt.show()
```

## Styling Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle='--', color='red')

plt.show()
```

```python
plt.plot(x, y, marker='s', markersize=10, markerfacecolor='green')

plt.show()
```

```python
plt.plot(x, y, marker='p', color='slateblue', ms=17)

plt.show()
```

## Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales)

plt.show()
```

```python
plt.bar(products, sales, color='skyblue', width=0.5)

plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
size = np.random.rand(50) * 100
color = np.random.rand(50)

plt.scatter(x, y, s=size, c=color, cmap='viridis')

plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

## Subplots

```python
x = np.linspace(0,10,100)
y = np.sin(x)
data = np.random.randn(1000)

fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x, y)
ax2.hist(data, edgecolor='black')

ax1.set_title('Sine Wave')
ax2.set_title('1000 Random Numbers')

ax1.set_xlabel('X-axis')
ax2.set_xlabel('X-axis')

ax1.set_ylabel('Y-axis')
ax2.set_ylabel('Y-axis')
plt.tight_layout()
plt.show()
```
