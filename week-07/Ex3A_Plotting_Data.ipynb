# Exercise 3.A Getting Familiar with Matplotlib

```python

import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title('Top of graph!')
plt.xlabel('Bottom of graph')
plt.ylabel('Left of graph')
plt.show()
```

## Styling Plots

```python
plt.plot(x, y, marker='o')
plt.title('Top of graph!')
plt.xlabel('Bottom of graph')
plt.ylabel('Left of graph')
plt.show()
```
