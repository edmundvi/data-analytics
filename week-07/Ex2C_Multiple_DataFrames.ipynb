## Basic Aggregations

```python
data = {'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East',
'West', 'North'],
'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]}

df_sales = pd.DataFrame(data)

df_sales
```

```python
df_sales['Revenue'] = df_sales['Units_Sold'] * df_sales['Price']

df_sales
```

```python
print(f'Sum of Units_Sold: {df_sales['Units_Sold'].sum()}')

print(f'Sum of Revenue: {df_sales['Revenue'].sum()}')
```

```python
print(f'Mean of Units_Sold: {df_sales['Units_Sold'].mean().round(2)}')
print(f'Mean of Revenue: {df_sales['Revenue'].mean().round(2)}')
#learned how to use round with sum and mean from in class!
```
