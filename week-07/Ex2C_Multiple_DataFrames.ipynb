# Exercise 2.C Working with Multiple DataFrames

```python
import numpy as np
import pandas as pd
```

## Basic Aggregations

```python
data = {'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East',
'West', 'North'],
'Units_Sold': [10, 15, 20, 5, 10, 15, 10, 5, 15],
'Price': [100, 200, 150, 100, 200, 150, 100, 200, 150]}

df_sales = pd.DataFrame(data)

df_sales
```

```python
df_sales['Revenue'] = df_sales['Units_Sold'] * df_sales['Price']

df_sales
```

```python
print(f'Sum of Units_Sold: {df_sales['Units_Sold'].sum()}')

print(f'Sum of Revenue: {df_sales['Revenue'].sum()}')
```

```python
print(f'Mean of Units_Sold: {df_sales['Units_Sold'].mean().round(2)}')
print(f'Mean of Revenue: {df_sales['Revenue'].mean().round(2)}')
#learned you could  use round with sum and mean from in class!
```

## Grouping Data

```python
df_sales.groupby('Product')[['Units_Sold', 'Revenue']].sum()
```

```python
df_sales.groupby('Region')[['Units_Sold', 'Revenue']].mean().round(2)
```

```python
df_sales.groupby(['Product', 'Region'])[['Units_Sold', 'Revenue']].sum()
```

## Sorting and Ranking Data

```python
df_sales.sort_values(by='Units_Sold')
```

```python
df_sales.sort_values(by='Revenue', ascending=False)
```

```python
df_sales['Revenue_Rank'] = df_sales['Revenue'].rank(ascending=False)

df_sales.sort_values(by='Revenue_Rank', ascending=False)
```

## Merging DataFrames

```python
product_data = {
    'Product': ['A', 'B', 'C'],
    'Category': ['Electronics', 'Furniture', 'Appliances']
}

df_products = pd.DataFrame(product_data)

df_products
```

```python
df_merged = pd.merge(df_sales, df_products, on='Product')

df_merged
```

```python
df_merged.groupby('Category')['Revenue'].sum()
```
