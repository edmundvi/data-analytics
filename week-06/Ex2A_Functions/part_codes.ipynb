# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings - part codes

```python
 def get_supplier_code(code):
    return code.split(':')[0]
```

```python
def get_product_num(num):
    return num.split(':')[1].split('-')[0]
```

```python
def get_size(size):
    return size.split('-')[1]
```

```python
sup1 = 'ACME:123-L'
sup2 = 'DI:12345-M'
sup3 = 'ACE:1-12'
```

```python
print(f'For {sup1}, the supplier is {get_supplier_code(sup1)}. Part number {get_product_num(sup1)} is available in size {get_size(sup1)}.')
```
