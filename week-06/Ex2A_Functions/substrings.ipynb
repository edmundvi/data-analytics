## Lab 1. Working with names - parsing strings

```python
test_name = 'John Doe'
```

```python
def parse_and_display(name):
    for i in range (len(name)):
        if name[i] == ' ':
            first = name [:i]
            last = name [i+1:]
            print (f'Full name: {name}')
            print (f'First name: {first}')
            print (f'Last name: {last}')
```

```python
parse_and_display(test_name)
```
